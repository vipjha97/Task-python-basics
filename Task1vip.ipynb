```python
# Sum of elements, greatest element & location of the same
import numpy as np       
# 1D array   
def arraysum(arr):
    add = np.sum(arr)
    return add   
arr1 = [20, 2, 0.2, 10, 4, 12, 14, 8, 9, 5] 
print("The sum is", arraysum(arr1))

def arraymax(arr):
    maxi= np.amax(arr)
    return maxi
print("The largest element is", arraymax(arr1))

def arraymaxloc(arr):
    maxiloc= np.where(arr == np.amax(arr))
    return maxiloc
print("The position of largest element", arraymaxloc(arr1))
```

```python
import cv2 
import numpy as np 
try: 
    # Read image from the disk. 
    img = cv2.imread(r"C:\Users\Vipul Jha\Desktop\flower.png") 
  
    # Shape of image in terms of pixels. 
    (rows, cols) = img.shape[:2] 
  
    # getRotationMatrix2D creates a matrix needed for transformation. 
    # We want matrix for rotation w.r.t center to 45 degree without scaling. 
    M = cv2.getRotationMatrix2D((cols / 2, rows / 2), 45, 1) 
    res = cv2.warpAffine(img, M, (cols, rows)) 
  
    # Write image back to disk. 
    cv2.imwrite('result.jpg', res) 
except IOError: 
    print ('Error while reading files !!!') 


        


```
